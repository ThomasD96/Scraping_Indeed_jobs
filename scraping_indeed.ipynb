- On can now write functions to extract these pieces of useful information from a soup object of a job listing. The functions below extract the job title, company name, company rating, no. of reviews, and the job description. These will come in handy when we automate the process of querying hundreds of job listings.

```python
def get_job_title(soup):
    return soup.find(id="indeed-share-message")['content']

def get_company(soup):  
    ''' Input: Soup object
        Returns: company name (if found), else None
    '''
    import re
    try:
        company_name = soup.find(class_="jobsearch-CompanyReview--heading").get_text()
        return company_name
    except AttributeError:        
        footer_links = soup.find_all(class_="jobsearch-RelatedLinks-link")
        for el in footer_links:
            if el.get_text().startswith('Jobs at'):
                company_name = re.search('at(.*)in', el.get_text()).group(1).strip()
                return company_name
    except AttributeError:
        company_name = soup.find(class_="jobsearch-CompanyAvatar-cta").get_text().split('Get job updates from ')[1].strip()
        return company_name
    except:
        return None
    


def get_num_reviews(soup):
    '''Returns the review count (if available) of the company on Indeed given a soup object.'''
    try:
        reviews_count = soup.find(class_ = "icl-Ratings-count").get_text()
        return reviews_count
    except:
        return np.nan

def get_company_rating(soup):
    '''
    Argument: soup object
    Returns: the compnay rating (if available) of the company.
    '''    
    try:
        rating = soup.find(class_="icl-Ratings-starsCountWrapper")['aria-label'].split()[0]
        return float(rating)
    except:
        return np.nan
    
from nltk.corpus import stopwords

def get_job_desc(soup):
    ''' 
    Argument: Soup object
    Returns: job description as a list of words after filtering out English stopwords
    '''
    job_desc = soup.find(id="jobDescriptionText").get_text()
    job_desc = job_desc = [word.strip(',.:')  for word in job_desc.lower().split()]   
    en_stops = set(stopwords.words('english'))
    list_of_words = list(set(job_desc) - en_stops)
    return list_of_words

def scrape_job_listing(url):
    '''
    Argument: An url of a particular job listing.
    Returns: After scraping and parsing the webpage, it returns 
    the company name, job title, company rating, and job description as a tuple.
    '''
    try:
        response = requests.get(url)
        response.raise_for_status()
    except requests.exceptions.HTTPError as err:    
        print(err)
        sys.exit(1)
    except requests.exceptions.RequestException as err:
        print(err)
        sys.exit(1)
    soup = BeautifulSoup(response.content, 'lxml')
    num_reviews = get_num_reviews(soup)
    job_title = get_job_title(soup)
    company_name = get_company(soup)
    job_desc = get_job_desc(soup)
    company_rating = get_company_rating(soup)
    return (company_name, job_title, company_rating, job_desc)
```

### Now it's time to bring it all together. 
 - Define functions to create a search url from 1.) the base url and 2.) desired job location and job title. Read and parse the contents of the search results.
 - Iterate over the pages of the search results and for each page, read all the job links and parse specific information about the jobs.

```python
def make_url(base_url,job_title,city):
    '''
    Input: base_url (country-specific), job title and the city
    Returns: search query url
    '''
    query_url = base_url+'/jobs?q='+job_title+'&l='+city
    return query_url

def get_page_url(user_query_url, page_num):
    '''
    Returns the the url of a particular page of the search query
    '''
    return user_query_url+'&start={}'.format((page_num)*10)

def soupify(query_url):
    '''
    Takes the current page
    '''
    search_response = requests.get(query_url).content
    search_soup = BeautifulSoup(search_response, 'lxml')
    return search_soup


def scrape_current_page(page_url, scrape_ads):
    jobs_data_curr_page = []
    search_soup = soupify(page_url)
    if scrape_ads: # if you want to scrape the sponsored listings too
        job_links = [base_url+job.get('href') for job in search_soup.find_all(class_="jobtitle turnstileLink")]
    else:
        job_links = [base_url+job.get('href') for job in search_soup.find_all(class_="jobtitle turnstileLink") if not job.get('href').startswith('/pagead/')]    
    for job_link in job_links:
        data = scrape_job_listing(job_link)
#        print('Scraped job at {}'.format(data[0]))
        jobs_data_curr_page.append(data)
        sleep(1)
    return jobs_data_curr_page


def scrape_pages(base_url, desired_job_title='data+scientist', desired_location='', scrape_ads=True):
# First set up the search query url based on     
    user_query_url = make_url(base_url,desired_job_title,desired_location)
    search_soup = soupify(user_query_url)
    num_jobs = int(search_soup.find(id = 'searchCount').contents[0].split()[-2])
    if scrape_ads:
        num_pages = int(np.ceil(num_jobs/15))
    else:
        num_pages = int(np.ceil(num_jobs/10))
    full_data = pd.DataFrame(columns=['company_name', 'job_title', 'company_rating','job_description'])
    
    for page_num in range(num_pages):
        print('Scraping page number {}'.format(page_num+1))
        page_url = get_page_url(user_query_url, page_num)
        page_data = scrape_current_page(page_url, scrape_ads)
        clean_page_data_df = pd.DataFrame(page_data, columns=['company_name', 'job_title', 'company_rating','job_description'])
        full_data = pd.concat([full_data,clean_page_data_df],axis=0, ignore_index=True)
    print('Done! Scraped {} jobs from {} pages'.format(len(full_data), num_pages))
    return full_data
```

```python
def visualise_skills_required(job_descs_word_freq,city):
    '''
    Input: Word counts as a Counter object
    Output: Bar plot of different keywords segemented by groups
    '''
    
    prog_language_keywords_dict = { 
                                'R':job_descs_word_freq['r'], 'Python':job_descs_word_freq['python'],
                               'Java':job_descs_word_freq['java'], 'C/C++':job_descs_word_freq['c']+job_descs_word_freq['c++'],                     
                               'Ruby':job_descs_word_freq['ruby'], 'Weka':job_descs_word_freq['weka'],
                               'Perl':job_descs_word_freq['perl'], 'Matlab':job_descs_word_freq['matlab'],
                               'JavaScript':job_descs_word_freq['javascript'], 'Scala': job_descs_word_freq['scala'],
                               'Shell/bash': job_descs_word_freq['shell']+job_descs_word_freq['bash']
                              }

    big_data_keywords_dict = {'Hadoop':job_descs_word_freq['hadoop'], 'MapReduce':job_descs_word_freq['mapreduce'],
                              'Spark':job_descs_word_freq['spark'], 'Pig':job_descs_word_freq['pig'],
                              'Hive':job_descs_word_freq['hive'], 'AWS':job_descs_word_freq['aws'],
                              'Heroku': job_descs_word_freq['heroku'],'Azure':job_descs_word_freq['azure'],
                              'Yarn':job_descs_word_freq['yarn']                                          
                             } 

    database_keywords_dict = {
                              'SQL':job_descs_word_freq['sql'], 'NoSQL':job_descs_word_freq['nosql'],
                              'HBase':job_descs_word_freq['hbase'], 'Cassandra':job_descs_word_freq['cassandra'],
                              'MongoDB':job_descs_word_freq['mongodb'], 'Impala':job_descs_word_freq['impala']                                    
                             }

                      
    analysis_tool_dict = {
                          'Excel':job_descs_word_freq['excel'],  'Tableau':job_descs_word_freq['tableau'],
                          'SAS':job_descs_word_freq['sas'], 'Bokeh':job_descs_word_freq['bokeh'],
                          'SPSS':job_descs_word_freq['spss'], 'Shiny':job_descs_word_freq['shiny'],
                          'd3.js': job_descs_word_freq['d3.js']+job_descs_word_freq['d3'],
                          'ggplot': job_descs_word_freq['ggplot']
                         } 


    devops_dict = {
                        'Git/Github': job_descs_word_freq['git']+job_descs_word_freq['github'], 
                       'Jenkins': job_descs_word_freq['jenkins'], 'Docker': job_descs_word_freq['docker'],
                        'Kubernetes': job_descs_word_freq['kubernetes'],           
                       'Puppet': job_descs_word_freq['puppet']
                   }

    python_packages_dict = {
                    'Numpy': job_descs_word_freq['numpy'], 'Scipy': job_descs_word_freq['scipy'],
                    'Tensorflow':job_descs_word_freq['tensorflow'], 'Pandas': job_descs_word_freq['pandas'],
                    'Matplotlib': job_descs_word_freq['matplotlib'], 'Seaborn': job_descs_word_freq['seaborn'],                 
                     'scikit-learn': job_descs_word_freq['scikit'] + job_descs_word_freq['scikit-learn']+job_descs_word_freq['sklearn'],
                     'Pytorch':job_descs_word_freq['pytorch'], 'Keras': job_descs_word_freq['keras'] 
                            }

    keywords_dict_combined = {**prog_language_keywords_dict, **big_data_keywords_dict,
                          **database_keywords_dict, **analysis_tool_dict, 
                          **devops_dict, **python_packages_dict}
    
    prog_language_keywords_df = make_series_skills(prog_language_keywords_dict)
    big_data_keywords_df      = make_series_skills(big_data_keywords_dict)
    database_keywords_df      = make_series_skills(database_keywords_dict)
    analysis_tool_df          = make_series_skills(analysis_tool_dict)
    devops_df                 = make_series_skills(devops_dict)
    packages_df               = make_series_skills(python_packages_dict)

    fig, axs = plt.subplots(3, 2, figsize=(18, 16),tight_layout=True)
    sns.barplot(x=prog_language_keywords_df.index, y = prog_language_keywords_df/num_jobs_scraped * 100, orient='v',ax=axs[0][0])
    sns.barplot(x=big_data_keywords_df.index, y = big_data_keywords_df/num_jobs_scraped * 100, orient='v',ax=axs[0][1])
    sns.barplot(x=database_keywords_df.index, y = database_keywords_df/num_jobs_scraped * 100, orient='v',ax=axs[1][0])
    sns.barplot(x=analysis_tool_df.index, y = analysis_tool_df/num_jobs_scraped * 100, orient='v',ax=axs[1][1])
    sns.barplot(x=packages_df.index, y = packages_df/num_jobs_scraped * 100, orient='v',ax=axs[2][0])
    sns.barplot(x=devops_df.index, y = devops_df/num_jobs_scraped * 100, orient='v',ax=axs[2][1])

    from itertools import chain

    for ax in list(chain.from_iterable(zip(*axs))):
        ax.set_ylabel('')
    
    axs[0][0].set_title('Programming languages', fontsize='large')
    axs[0][1].set_title('Big Data technologies', fontsize='large')
    axs[1][0].set_title('Databases', fontsize='large')
    axs[1][1].set_title('Visulalisation and analysis tools', fontsize='large')
    axs[2][0].set_title('Popular frameworks', fontsize='large')
    axs[2][1].set_title('DevOps', fontsize='large')
    _ = fig.text(0.0, 0.5, '%age of job listings', va='center', rotation='vertical', fontsize='x-large') 
    
    keywords_df = pd.DataFrame.from_dict(keywords_dict_combined, orient='index', columns=['num_occurences']).sort_values(by ='num_occurences',ascending=False)
    plt.figure(figsize=(17,12))
    _ = sns.barplot(y=keywords_df.index, x = keywords_df['num_occurences']/num_jobs_scraped * 100, orient='h')
    _ = plt.title('Essential skills in %s'%city)
    _ = plt.xlabel('Percentage of job listings that mention a certain skill')
    _ = plt.ylabel('Skill')
```

```python
def make_series_skills(keywords_dict):
    '''
    Input: Dictionary with keys = keywords, and values = frequency of that keyword
    Returns: A pandas Series with sorted values in descending order
    ''' 
    df = pd.Series(keywords_dict,name='num_occurences').sort_values(ascending=False)
    return df
```
